#### 2. Statistik Data

```python
df_a = data_negara_investor.groupby("nama_region").nunique()["nama_negara"]
df_a.sort_values(ascending=False)
```

```python
df_d = data_negara_investor.groupby("nama_region").sum()["nilai_investasi"]
df_d.sort_values(ascending=False)
```

```python
df_b = data_negara_investor.groupby(['nama_negara','tahun']).sum()['nilai_investasi']
df_b.sort_values(ascending=False)
```

#### 3. Visualisasi Data

```python
df_c = data_negara_investor.groupby('nama_region').sum()['nilai_investasi'].sort_values(ascending=False)
df_c = df_c.plot(kind='bar', title ="Wilayah Negara Investor", legend=True, fontsize=12)
df_c.set_xlabel("Negara", fontsize=12)
df_c.set_ylabel("Investasi", fontsize=12)
```

Dari tabel dan diagram di atas menerangkan bahwasannya negara-negara di Eropa banyak yang menjadi investor di Indonesia, namun demikian wilayah negara yang dengan nilai terbesar adalah Asia dan Singapura merupakan negara yang memiliki investasi terbesar di Indonesia.
